```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('wifi.json', lines=True)
copydata = data.copy()
copydata = copydata.drop("_id",1)
copydata = copydata.fillna(0)
```

```python
for column in copydata.columns:
    s = 0
    for i in copydata[column]:
        if i!=100:
            s+=1
    print(column,':',s) 
```

```python
new_wifi = copydata.filter(['D-Link_DIR-524','HP-Print-10-LaserJet Pro MFP','VLSIDESIGN_LAB','dlink','dlink_pksingh','x','y'],axis=1)
```

```python
new_wifi['cellid_'] = (new_wifi['x']-1)*47+(new_wifi['y'])
```

```python
columns_ = list(new_wifi.columns[:-3])
new_wifi.columns = ['AP01','AP02','AP03','AP04','AP05','x','y','cellid_']
new_wifi
```

new_wifi.columns = ['AP01':D-Link_DIR-524,'AP02':HP-Print-10-LaserJet Pro MFP,'AP03':VLSIDESIGN_LAB ,'AP04':dlink,'AP05':dlink_pksingh]

```python
for ap in columns_:
    nrrssr = ap+'-NrrssRight'
    nrrssl = ap+'-NrrssLeft'
    nrrsst = ap+'-NrrssTop'
    nrrssb = ap+'-NrrssBottom'

    new_wifi[nrrssr] = pd.Series(0, index=np.arange(new_wifi.shape[0]))
    new_wifi[nrrssl] = pd.Series(0, index=np.arange(new_wifi.shape[0]))
    new_wifi[nrrsst] = pd.Series(0, index=np.arange(new_wifi.shape[0]))
    new_wifi[nrrssb] = pd.Series(0, index=np.arange(new_wifi.shape[0]))

    for i, row in new_wifi.iterrows():
       
        for j, row2 in new_wifi.iterrows():
            if new_wifi['y'][i] == new_wifi['y'][j]:
                if new_wifi['x'][i] == new_wifi['x'][j]-1:
                    new_wifi[nrrssr][i] = new_wifi[ap][i] - new_wifi[ap][j]
            elif new_wifi['y'][i] == new_wifi['y'][j]:
                if new_wifi['x'][i] == new_wifi['x'][j]+1:
                    new_wifi[nrrssl][i] = new_wifi[ap][i] - new_wifi[ap][j]
            elif new_wifi['x'][i] == new_wifi['x'][j]:
                if new_wifi['y'][i] == new_wifi['y'][j]-1:
                    new_wifi[nrrssb][i] = new_wifi[ap][i] - new_wifi[ap][j]
            elif new_wifi['x'][i] == new_wifi['x'][j]:
                if new_wifi['y'][i] == new_wifi['y'][j]-1:
                    new_wifi[nrrsst][i] = new_wifi[ap][i] - new_wifi[ap][j]
                    
```

```python
new_wifi.to_csv('new_wifi.csv', index=False)
```
